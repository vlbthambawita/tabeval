## Generate Stratified Subsamples

### Subtask:
Create subsamples of sizes 1000, 800, 600, 400, and 200 from the `clean_data` DataFrame, ensuring that the proportional distribution of the 'Disease' column is maintained in each subsample. Store these in separate DataFrames.


**Reasoning**:
I need to import the `train_test_split` function from `sklearn.model_selection` to generate stratified subsamples as per the instructions.



```python
from sklearn.model_selection import train_test_split

print("Imported train_test_split from sklearn.model_selection.")
```

**Reasoning**:
Now that `train_test_split` is imported, I will define the subsample sizes, iterate through them, calculate the `test_size`, and use `train_test_split` with stratification on the 'Disease' column to create and store each subsample as instructed, printing a confirmation for each.



```python
# First, hold out a separate stratified test dataset (1000 samples)
TEST_SIZE = 1000
if TEST_SIZE > len(clean_data):
    raise ValueError(f"clean_data has {len(clean_data)} rows; cannot create test set of {TEST_SIZE} samples.")

train_data, test_data = train_test_split(
    clean_data,
    test_size=TEST_SIZE,
    stratify=clean_data['Disease'],
    random_state=42
)
print(f"Created separate test dataset with {len(test_data)} samples.")

# Generate stratified subsamples from the training data only (no overlap with test)
subsample_sizes = [1000, 800, 600, 400, 200]
stratified_subsamples = {}

print("Generating stratified subsamples from training data...")

for size in subsample_sizes:
    if size > len(train_data):
        print(f"Warning: Requested subsample size ({size}) is larger than train_data ({len(train_data)}). Skipping.")
        continue

    test_size_ratio = size / len(train_data)

    _, subsample = train_test_split(
        train_data,
        test_size=test_size_ratio,
        stratify=train_data['Disease'],
        random_state=42
    )

    stratified_subsamples[f'subsample_{size}'] = subsample
    print(f"Generated subsample_{size} with {len(subsample)} rows.")

print("Stratified subsamples generated successfully.")

```

**Reasoning**:
Now that the subsamples have been generated, I will verify the 'Disease' distribution in each subsample against the original `clean_data` to ensure stratification was successful. This involves calculating and comparing the proportions of each disease category.



```python
print("Verifying 'Disease' distribution in clean_data and subsamples...")

# Original distribution in clean_data
original_disease_distribution = clean_data['Disease'].value_counts(normalize=True).sort_index()
print("\n--- Original 'Disease' Distribution (clean_data) ---")
print(original_disease_distribution)

# Verify distribution in each subsample
for name, subsample_df in stratified_subsamples.items():
    subsample_disease_distribution = subsample_df['Disease'].value_counts(normalize=True).sort_index()
    print(f"\n--- '{name}' 'Disease' Distribution ---")
    print(subsample_disease_distribution)

    # Optional: Compare with original for verification
    print("Difference from original:")
    print(subsample_disease_distribution - original_disease_distribution)
```

**Reasoning**:

This involves consolidating all datasets, iterating through each column, calculating normalized value counts for each dataset, combining them into a single DataFrame, and then using seaborn to create a comparative bar plot as per the instructions.



```python
print("Generating comparative bar plots for each column...")

# Create a dictionary including clean_data and all subsamples
all_datasets = {'clean_data': clean_data}
all_datasets.update(stratified_subsamples)

# Iterate through each column in clean_data.columns
for col in clean_data.columns:
    plot_data_list = []
    for name, df in all_datasets.items():
        # Calculate normalized value counts
        value_counts = df[col].value_counts(normalize=True).reset_index()
        value_counts.columns = ['Category', 'Proportion']
        value_counts['Source'] = name # Add a 'Source' column to identify the dataset
        plot_data_list.append(value_counts)

    # Concatenate all DataFrames for the current column
    combined_df = pd.concat(plot_data_list)

    # Create the bar plot
    plt.figure(figsize=(12, 6))
    sns.barplot(data=combined_df, x='Category', y='Proportion', hue='Source', palette='viridis')
    plt.title(f'Comparative Distribution of {col} Across Datasets', fontsize=16)
    plt.xlabel('Category', fontsize=12)
    plt.ylabel('Proportion', fontsize=12)
    plt.xticks(rotation=45, ha='right') # Rotate x-axis labels for readability
    plt.legend(title='Dataset Source')
    plt.tight_layout() # Adjust layout to prevent labels from overlapping
    plt.show()

print("Comparative bar plots generated successfully.")
```

## Train Individual Synthesizers for Subsamples (N times)

### Subtask:
Iterate through each stratified subsample and train multiple `CTGANSynthesizer` models for that subsample. For each trained synthesizer, generate a small synthetic dataset of 100 rows to confirm functionality.


**Reasoning**:
The previous code was interrupted, and the instructions now specify to train *multiple* CTGANSynthesizer models for each subsample. I will implement the nested loop structure as described in the instructions, initializing dictionaries for each run and generating synthetic data accordingly.



```python
from sdv.single_table import CTGANSynthesizer

trained_synthesizers = {}
synthetic_data_samples = {}
num_training_runs = 3

print("Starting CTGAN synthesizer training for each subsample and multiple runs...")

for subsample_name, subsample_df in stratified_subsamples.items():
    print(f"\n--- Processing subsample: {subsample_name} ---")
    trained_synthesizers[subsample_name] = {}
    synthetic_data_samples[subsample_name] = {}

    for run in range(num_training_runs):
        print(f"  Training synthesizer for {subsample_name}, run {run + 1}/{num_training_runs}...")

        # Create a new instance of CTGANSynthesizer for each run
        # metadata was loaded previously and contains the table schema
        synthesizer = CTGANSynthesizer(metadata, verbose=True)

        # Train the synthesizer with the current subsample
        synthesizer.fit(subsample_df)

        # Store the trained synthesizer
        trained_synthesizers[subsample_name][f'run_{run}'] = synthesizer

        # Generate a small synthetic dataset (100 rows)
        synthetic_sample = synthesizer.sample(num_rows=100)

        # Store the synthetic data sample
        synthetic_data_samples[subsample_name][f'run_{run}'] = synthetic_sample

        print(f"  Synthesizer for {subsample_name}, run {run + 1} trained and 100 synthetic rows generated. First 5 rows:\n{synthetic_sample.head()}")

print("All CTGAN synthesizers trained and synthetic samples generated for all subsamples and runs.")
```
